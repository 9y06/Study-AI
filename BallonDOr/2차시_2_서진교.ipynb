```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/BallonDOr.csv', encoding='utf-8')

print(df.head())
print(df.columns)

# 수치 분석
average_points = df['Points'].mean()
print(f'평균 포인트: {average_points:.2f}')

winners_by_country = df['Nationality'].value_counts()
print('각 국가별 수상자 수:')
print(winners_by_country)

winners_summary = winners_by_country.describe()
print('수상자 수 통계 요약:')
print(winners_summary)

# 2. 데이터 시각화
# (a) 각 국가별 수상자 수를 막대그래프로 시각화
plt.figure(figsize=(12, 6))
winners_by_country.plot(kind='bar', color='skyblue')
plt.title('Number of Ballon d\'Or winners in each country')
plt.xlabel('country')
plt.ylabel('Number of winners')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('winners_by_country_bar_chart.png')  # 차트 저장
plt.show()

# (b) 연도별 수상자 수를 선그래프로 시각화
# 'Year' 열을 기준으로 수상자 수 집계
winners_by_year = df['Year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
winners_by_year.plot(kind='line', marker='o', color='orange')
plt.title('Ballon D\'Or Winners by Year')
plt.xlabel('Year')
plt.ylabel('Number of winners')
plt.grid()
plt.tight_layout()
plt.savefig('winners_by_year_line_chart.png')
plt.show()
```

평균 포인트는 약 23.92점입니다.
