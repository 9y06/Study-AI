# **과제 1번**
**K-평균 클러스터 알고리즘**을 사용하여 주어진 데이터셋을 학습해봅시다.


*   **데이터셋**: 맥도날드 고객 리뷰 데이터 1400개 (mcdonalds_data.csv)
*   **과제 내용**


1.   주어진 데이터셋을 K-평균 클러스터링 알고리즘을 생성해서 학습합니다.
2.   클러스터의 개수를 다르게 설정해가며 최적의 클러스터 개수를 찾고, 그 이유를 작성합니다.

````python
# 과제 1번 데이터 (맥도날드 고객 리뷰 데이터)
X = df
````

## 1. 모델 생성 및 학습
데이터셋을 K-평균 클러스터링 알고리즘을 생성하여 학습합니다. 클러스터의 개수는 5개로 설정합니다.


````python
# 모델 생성 및 학습
from sklearn.cluster import KMeans

model = KMeans(n_clusters = 5)
model.fit(X)
````

````python
# 데이터들의 클러스터 라벨 값을 확인해봅니다.
labels = model.labels_
print(labels)
````

## 2. 클러스터 특징 분석
같은 클러스터에 속한 데이터끼리 비슷한 특징을 가지는지 확인해봅니다.   
판다스 라이브러리의 다양한 함수들을 활용해서 분석을 해봅니다.

````python
# 같은 클러스터에 해당하는 데이터끼리 묶기(실행만 하면 됩니다!)
clusters = []
for i in range(5):
    clusters.append(df[model.labels_ == i])
````

````python
# 원하는 클러스터에 속하는 데이터 불러오기 (클러스터는 0부터 4까지 존재)
num = int(input("원하는 클러스터 번호를 입력하세요 (0 ~ 4): "))
df_cluster = clusters[num]
````

````python
# 판다스 라이브러리를 활용하여 특징 분석하기
# 원하는 클러스터의 상관관계 계산
print(f"클러스터 {df_cluster}의 상관관계:")
print(clusters[num].corr())
````

## 2. 최적의 클러스터 개수 찾기
클러스터의 개수를 다르게 설정해가며 최적의 클러스터 개수를 찾아봅니다.

````python
# 최적의 클러스터 개수 찾기
model = KMeans(n_clusters=4, random_state=45)
model.fit(X)
````

## 1. 환경 불러오기
선택한 강화학습 환경을 불러옵니다.

````python
# 강화학습 환경 불러오기
import gym

env = gym.make('CartPole-v1')
````

## 2. 강화학습 알고리즘 학습
강화학습 코드를 사용하여 선택한 환경을 학습합니다.


````python
# 강화학습 코드
num_episode = 20

# 아래의 for문을 완성해주세요.
for i_episode in range(num_episode):
    # 강화학습 환경을 초기화합니다
    state = env.reset()
    total_reward = 0  # 누적 보상 초기화

    # 하나의 에피소드동안 총 100번의 타임스텝이 진행됩니다
    for t in range(100):

        # 강화학습 환경을 렌더링합니다
        env.render()

        # 이번 타임스텝에 수행할 액션을 선택합니다
        action = env.action_space.sample()

        # 환경의 상태 변화 및 보상을 계산합니다
        next_state, reward, done, _ = env.step(action)
        total_reward += reward

        # 종료 조건이 만족될 경우 에피소드를 종료합니다
        if done:
            print(f"에피소드 {i_episode + 1} 종료 - 총 타임스텝: {t + 1}, 총 보상: {total_reward}")
            break

# 강화학습 환경을 종료합니다
env.close()
````

## 3. 시각화
아래의 시각화 코드를 사용하여 학습한 강화학습 환경을 렌더링합니다.

````python
# 렌더링을 위해 반드시 설치해 주세요.
!apt update
!apt install xvfb
!pip install pyvirtualdisplay
````

````python
# 에피소드 시각화

# 렌더링에 필요한 라이브러리 불러오기
from IPython import display as ipythondisplay
from pyvirtualdisplay import Display
from PIL import Image

# 각 에피소드를 렌더링하는 함수를 정의합니다
def render_episode(env):

    # 강화학습 환경을 렌더링할 스크린을 생성합니다
    prev_screen = env.render(mode='rgb_array')

    # 렌더링한 스크린을 이미지로 변환하여 배열에 저장합니다
    images = [Image.fromarray(prev_screen)]

    # 강화학습 환경을 초기화합니다
    obs = env.reset()

    # 총 100번의 타임스텝동안 학습을 반복합니다
    for t in range(100):

        # 액션을 선택합니다
        action = env.action_space.sample()

        # 환경의 상태 변화 및 보상을 계산합니다
        obs, reward, done, info = env.step(action)

        # 짝수번쨰의 타임스텝일 때마다 진행상황을 렌더링합니다
        if t % 2 == 0:
            screen = env.render(mode='rgb_array')

            # 해당 타임스텝을 렌더링한 스크린을 이미지로 변환하여 배열에 저장합니다
            images.append(Image.fromarray(screen))

        # 종료 조건이 만족될 경우 에피소드를 종료합니다
        if done:
            break

    return images
````

````python
import gym
from PIL import Image

# 렌더링 GIF를 생성하기 위한 함수를 정의합니다
def render_episode(env):
    images = []
    state = env.reset()
    done = False

    while not done:
        image = env.render(mode='rgb_array')
        images.append(Image.fromarray(image))
        action = env.action_space.sample()
        state, reward, done, info = env.step(action)

    return images

# 선택한 환경을 입력합니다
env = gym.make('CartPole-v1')

# 렌더링 GIF를 생성하고 저장합니다
images = render_episode(env)
images[0].save(
    'reinforcement_learning.gif', save_all=True, append_images=images[1:],
    loop=0, duration=100)  # duration은 프레임 간의 시간 간격 (ms)

# 강화학습 환경을 종료합니다
env.close()
````
